```python
import pandas as pd
import numpy as np 

from sklearn.preprocessing import LabelEncoder
```

```python
trip_reason = pd.read_csv('../data/TripReason.csv')
```

```python
trip_reason.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train , test = train_test_split(trip_reason, test_size=.3, random_state=823, stratify=trip_reason.Cancel)
```

```python
train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
```

```python
train
```

```python
test
```

```python
ans = test[['TripReason']]
```

```python
test.drop('TripReason', axis=1, inplace=True)
test
```

```python
train.to_csv('../data/train_data.csv', index=False)
test.to_csv('../data/test_data.csv', index=False)
ans.to_csv('../data/df_true.csv', index=False)
```
