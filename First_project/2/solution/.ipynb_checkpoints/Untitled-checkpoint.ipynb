```python
selected_cols = ['Created', 'DepartureTime', 'BillID', 'TicketID', 'ReserveStatus', \
                'Male', 'Price', 'CouponDiscount', 'From', 'To', 'Domestic',\
                'VehicleClass', 'Vehicle', 'Cancel', 'BuyerMobile', 'TripReason']


df = trip_reason[selected_cols]
df
```

```python
df[['Created', 'DepartureTime']] = df[['Created', 'DepartureTime']].apply(pd.to_datetime)
```

```python
df['TimeDiff'] = (df['DepartureTime'] - df['Created']).dt.days
df
```

```python
df.BillID.nunique() , df.TicketID.nunique()
```

```python
df['TicketPerOrder'] = df.groupby('BillID').TicketID.transform('count')
df
```

```python
sample = df.sample(5).Male

sample
```

```python
True in sample.values and False in sample.values
```

```python
def is_family (rows) : 
    return True in rows.values and False in rows.values    
```

```python
df['is_with_familiy'] = df.groupby('BillID').Male.transform(is_family)
df
```

```python
df['net_price'] = df.Price - df.CouponDiscount
df
```

```python
df['has_discount'] = df.CouponDiscount > 0
```

```python
df
```

```python
cities = df.From.append(df.To).values 
city_encoder = LabelEncoder().fit(cities)

df['From_encoded'] = city_encoder.transform(df.From)
df['To_encoded'] = city_encoder.transform(df.To)
df
```

```python
df.isna().sum() / df.shape[0]
```

```python
features = ['ReserveStatus',
       'Male', 'Domestic', 'Vehicle', 'Cancel', 'TripReason',
       'TimeDiff', 'TicketPerOrder', 'is_with_familiy', 'net_price',
       'From_encoded', 'To_encoded', 'has_discount']

data= df[features]
```

```python
data
```

```python
cat_cols = data.select_dtypes(exclude=['int','float']).columns
cat_les = {}

for col in cat_cols : 
    le = LabelEncoder().fit(data[col])
    cat_les.update({col:le})
    
cat_les
```

```python
for col in cat_cols :
    data[col] = cat_les[col].transform(data[col])

data
```

```python
data['net_price'] /= data['net_price'].max()
data
```

```python
from sklearn.model_selection import train_test_split

target_col = 'TripReason'
x_train,x_test , y_train,y_test = train_test_split(data.drop(target_col,axis=1), data[target_col], test_size=.2,\
                                                   random_state=535, stratify=data[target_col])
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()

model.fit(x_train, y_train)
```

```python
import seaborn as sns
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))

sns.heatmap(confusion_matrix(y_pred, y_test), annot=True)
```
