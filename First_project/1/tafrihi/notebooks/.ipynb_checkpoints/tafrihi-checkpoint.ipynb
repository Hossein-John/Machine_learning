```python
train_data['Created'] = pd.to_datetime(train_data['Created'])
train_data['DepartureTime'] = pd.to_datetime(train_data['DepartureTime'])
```

```python
train_data['Created_Month'] = train_data['Created'].dt.month
train_data['Created_DayOfWeek'] = train_data['Created'].dt.dayofweek

train_data['Departure_Month'] = train_data['DepartureTime'].dt.month
train_data['Departure_DayOfWeek'] = train_data['DepartureTime'].dt.dayofweek
```

```python
train_data.head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
```

```python
train_data['Created_Season'] = train_data['Created_Month'].apply(get_season)
train_data['Departure_Season'] = train_data['Departure_Month'].apply(get_season)
```

```python
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)
            
submission.to_csv('submission.csv', index=False)
file_names = ['tafrihi.ipynb', 'submission.csv']
compress(file_names)
```
