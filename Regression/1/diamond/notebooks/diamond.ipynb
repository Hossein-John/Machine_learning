```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
x1 = OrdinalEncoder(categories = [['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']], dtype = np.int64)
Test['cut'] = x1.fit_transform(Test[['cut']])
```

```python
Test.head()
```

```python
x2 = OrdinalEncoder(categories = [['J','I','H','G','F','E','D']], dtype = np.int64)
Test['color'] = x2.fit_transform(Test[['color']])
```

```python
Test.head()
```

```python
x3 = OrdinalEncoder(categories = [['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']],dtype=np.int64)
Test['clarity'] = x3.fit_transform(Test[['clarity']])
```

```python
Test.head()
```

```python
# predict test samples

y_pred_test = model.predict(Test)
submission = pd.DataFrame(y_pred_test, columns = ['price'])
submission
```

```python
import zipfile
import joblib

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)
joblib.dump(model, 'model')
file_names = ['diamond.ipynb', 'submission.csv', 'model']
compress(file_names)
```
